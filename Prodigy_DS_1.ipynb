```python
df.drop(['2023'],axis=1,inplace=True)
df
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df['Country Name'].unique()
```

```python
countries=['United Kingdom','India','China','United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']
data=df[df['Country Name'].isin(countries)]
plt.bar(data['Country Name'],data['2022'])
plt.xlabel("Country Name")
plt.ylabel("Population in 2022")
plt.title("Population of Countries in 2022")
plt.xticks(rotation=45)
plt.show()
```

```python
total_population=df[df['Indicator Code']=='SP.POP.TOTL']
```

```python
total_pop_sort=total_population.sort_values(by='2022',ascending=False)
total_pop_sort
```

```python
highest_total_pop=total_pop_sort.head()
highest_total_pop
```

```python
plot=sns.barplot(x='2022',y='Country Name',data=highest_total_pop)
plt.title("TOTAL POPULATION OF TOP 10 COUNTRIES IN 2022")
plt.xlabel("POPULATION")
plt.show()
```

```python
df['Average Growth Rate']=df.loc[:,'2012':'2022'].mean(axis=1)
```

```python
plt.figure(figsize=(12,8))
plt.barh(df['Country Name'].head(10),df['Average Growth Rate'].head(10))
plt.xlabel("Average Growth Rate")
plt.ylabel("Country")
plt.show()
```
